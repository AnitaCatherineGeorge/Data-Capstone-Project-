# Exploratory Data Analysis

**Handling Missing values (if any) and using heatmap to visualize null values**

```python
df.isna().sum()
```

```python
sns.heatmap(df.isnull(), yticklabels = False)
```

**Absence hours per reason**

```python
plt.figure(figsize = (16, 6))
sns.boxplot(x = df['Reason for absence' ], y = df['Absenteeism time in hours'])
plt.show()
```

What one can see in the boxplot:

Reason 9 (Diseases of the circulatory system) causes longer absences, followed by Reason 12 (Diseases of the skin and subcutaneous tissue).

**Look for the season that had the highest number of absenteesm**

```python
plt.figure(figsize=(7,7))
X=df.groupby(['Seasons'])['Absenteeism time in hours'].sum().index
Y=df.groupby(['Seasons'])['Absenteeism time in hours'].sum()
clrs = ['indigo' if (x < max(Y)) else 'red' for x in Y ]
import seaborn as sb
sb.barplot(x=X, y=Y, palette=clrs)
        
# Show the plot
sb.set()
```

Season 3 had the highest employee absence.

**Top 5: Most absence hours per observation**

```python
df1 = pd.DataFrame(df, columns= [ 'ID','Month of absence',  'Reason for absence', 'Absenteeism time in hours'])

# sort 'Absenteeism time in hours' - ascending order
df1.sort_values(by=['Absenteeism time in hours', 'ID'], ascending = False, inplace = True)

print(df1.nlargest(5, ['Absenteeism time in hours']))
```

Display the details of the top 5 absentees

```python
df.loc[[323, 420,729,471,622]] 
```

**Leat 5: Least absence hours per observation**

```python
df2 = pd.DataFrame(df, columns= ['ID', 'Month of absence',  'Reason for absence', 'Absenteeism time in hours'])

# sort 'Absenteeism time in hours' - ascending order
df2.sort_values(by=['Absenteeism time in hours','ID'], ascending = 'True', inplace = True)

print(df2.nsmallest(5, ['Absenteeism time in hours']))
```

Display the details of the least 5 absentees

```python
df.loc[[405,203,714,446,737]]
```

**Employees without any absence**

**Look for the month and weekday that had the highest no of absences**

```python
plt.figure(figsize = (16,6))
plt.subplot(1, 2, 1)

sns.barplot(df.groupby(['Day of the week'])['Absenteeism time in hours'].sum().index, 
            df.groupby(['Day of the week'])['Absenteeism time in hours'].sum(), color = 'indigo')
plt.subplot(1, 2, 2)
sns.barplot(df[df['Month of absence'] != 0].groupby(['Month of absence'])['Absenteeism time in hours'].sum().index, 
            df[df['Month of absence'] != 0].groupby(['Month of absence'])['Absenteeism time in hours'].sum(), color = 'red')
plt.show()
```

Monday saw the highest absentees overall and March saw the highest absence in 3 years of survey.

**Heatmap to see positive/negative correlation between variables**

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True, fmt='.0%')
```

**Check out for the month in a particular season which had the highest absenteesm**

```python
plt.figure(figsize = (15, 6))
sns.heatmap(df[df['Month of absence'] != 0].groupby(['Seasons', 'Month of absence'])['Absenteeism time in hours'].sum().unstack()[0:], 
            annot = True, fmt = 'g', linewidths=.5, cmap="YlGnBu") 
plt.show()
```

Here Season 3 has the highest absentees. But the month that has highest absentees comes under Season 1 and it is in July.

**Check out for the weekday in a particular month which had the highest absenteesm**

```python
plt.figure(figsize = (15, 6))
sns.heatmap(df.groupby(['Month of absence', 'Day of the week'])['Absenteeism time in hours'].sum().unstack()[1:13], 
            annot = True, fmt = 'g',linewidths=.5, cmap="YlGnBu") 
plt.show()
```

**Looks like Tuesday in July had the highest absenteesm**

 **Check out for the reason for absence in that particular month and look for the highest reason**

```python
plt.figure(figsize = (15, 6))
sns.heatmap(df[df['Reason for absence'] != 0].groupby(['Month of absence', 'Reason for absence'])['Absenteeism time in hours'].sum().unstack(), 
            linewidths=.5, cmap="YlGnBu", annot = True, fmt = 'g')
plt.show() 
```

Reason 13 (Diseases of the musculoskeletal system and connective tissue ) caused mosed absenteesm over the years and highest case was in April.

 **Total no of employees who were absent in a particular day**

```python
plt.figure(figsize = (15, 6))
sns.heatmap(df.groupby(['Month of absence', 'Day of the week'])['ID'].nunique().unstack()[1:13], linewidths=.5, cmap="YlGnBu", annot = True, fmt = 'g')
plt.show()
```

Wednesday in the month of May saw more employees not making to work.

**Check out the total hours of absence each employee took in a particular month**

```python
plt.figure(figsize = (15, 6))
sns.heatmap(df.groupby(['Month of absence', 'ID'])['Absenteeism time in hours'].sum().unstack()[1:13], 
            linewidths=.5, cmap="YlGnBu", annot = True, fmt = 'g')
plt.show() 
```

Employee ID No 4,8 and 35 can be seen taking no leave. It can be inferred that they may be new employees.

```python
df['Work load Average/day '] = df['Work load Average/day ']/24
print(df['Work load Average/day '].unique())
```

**Check out the work load of employees with different education background**

```python
sns.factorplot(x='Day of the week', y='Work load Average/day ',data = df, hue='Education', col='Education', kind='swarm')
```

It is clear that employees with Education=1 background had excessive workload throughout the week.

**Check out the absence taken by employees with different education background**

```python
sns.factorplot(x='Day of the week', y='Absenteeism time in hours',data = df, hue='Education', col='Education', kind='swarm')
```

It is clear that employees with Education=1 background had taken the most leaves.

```python
print(df.groupby(['Social drinker','Education'])['Absenteeism time in hours'].describe())
```

**Check out which educated employees were the Social drinkers**

```python
sns.barplot(x = 'Education',
            y = 'Absenteeism time in hours',
            hue = 'Social drinker',
            data = df,
            ci = 0,
            palette = "Blues")
plt.ylabel('Mean of Absence in hrs')
plt.show()
```

Employees with Education = 1 background tended more to be social drinkers and they took more absence. Employees with Education=3 didn't have drinking problem.

**Check out which educated employees were the Social drinkers**

```python
sns.barplot(x = 'Education',
            y = 'Absenteeism time in hours',
            hue = 'Social smoker',
            data = df,
            ci = 0,
            palette = "Blues")
plt.ylabel('Mean of Absence in hrs')
plt.show()
```

Employees with Education = 1 background tended more to be social smokers and they took more absence. Employees with Education=3 didn't have smoking problem.

**Check out which educated employees had the most 'Discplinary failure' incidents**

```python
sns.barplot(x = 'Education',
            y = 'Service time',
            hue = 'Disciplinary failure',
            data = df,
            ci = 0,
            palette = "Blues")
plt.ylabel('Mean of Service Time')
plt.show()
```

Here it is clear that employees with Education=3 had the most service time and disciplinary issues but they are only 3 in number (number of employees). So lets conclude that employees with Education = 1 who are large in number had more service time and disciplinary issues.

**Check out which employee belonged to which education category, took more absence and had the drinking problem**

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(data=df, x="ID", y="Absenteeism time in hours", hue="Education",  palette="deep", style="Social drinker")
x=df['ID'].unique()
plt.xticks(x)
plt.show()
```

It is interesting to note that ID = 11,13,14,15,36 has the drinking problem and has taken more leaves than anyone else.

**Check out which employee belonged to which education category, took more absence and had the disciplinary issues**

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(data=df, x="ID", y="Absenteeism time in hours", hue="Education",  palette="deep", style="Disciplinary failure")
x=df['ID'].unique()
plt.xticks(x)
plt.show()
```

It is interesting to note that ID = 11,13,15,36 had the disciplinary issues at work and has taken more leaves.

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(data=df, x="ID", y="Service time", hue="Education",  palette="deep")
x=df['ID'].unique()
plt.xticks(x)
plt.show()
```

Education = 1 has the most highest 'Service time' average followed by Education = 2
